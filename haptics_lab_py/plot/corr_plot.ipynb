# サンプルデータのロード

```python
import plot_style
import seaborn as sns

df = sns.load_dataset('titanic')
```

# 全体の相関をプロット

```python
sns.pairplot(data=df, hue='alive')
```

# ヒートマップの作成

```python
sns.heatmap(data=df[['survived', 'pclass', 'age', 'sibsp','parch']].corr())
```

# ジョイントプロットの作成

```python
sns.jointplot(data=df, x='age', y='fare', kind='reg')
```
