# サンプルデータのロード

```python
import haptics_lab_py.plot.plot_style as plot_style
import seaborn as sns

df = sns.load_dataset('titanic')
```

# 箱ひげ図（ボックスプロット）の作成

```python
g = sns.catplot(data=df, x="class", y="age", kind="box", hue="alive")

# ラベルを変更
g.set_xlabels("Class")
g.set_ylabels("Age")
# 凡例を変更
g._legend.set_title("Alive")
for t, l in zip(g._legend.texts, ["No", "Yes"]):
    t.set_text(l)
```

# スウォームプロットの作成

```python
sns.catplot(data=df, x="class", y="age", kind="swarm", hue="alive")
```

# ヴァイオリンプロットの作成

```python
sns.catplot(data=df, x="class", y="age", kind="violin", hue="alive")
```

# ボックス＋スウォーム

```python
g = sns.catplot(data=df, x="class", y="age", hue="alive", kind='swarm', dodge=True, linewidth=1, palette="Set2")
g.map(sns.boxplot, data=df, x="class", y="age", hue="alive", dodge=True, palette="Set2")

g.set_xlabels("Class")
g.set_ylabels("Age")
g._legend.set_title("Alive")
for t, l in zip(g._legend.texts, ["No", "Yes"]):
    t.set_text(l)
```
